```python
_LOG_DIR = './betterhome_recipe'
```

```python
def read_json_strings(filepath):
    with open(filepath, 'r', encoding='utf-8') as r:
        data = json.load(r)
    return data
```

```python
def fetch_nest_dictkey(dic):
    for key, value in dic.items():
        yield key
        if isinstance(value, dict):
            yield from fetch_nest_dictkey(value)
```

```python
json_col = []
count = 0
for f in os.listdir(_LOG_DIR):
    if os.path.isdir(os.path.join(_LOG_DIR, f)):
        pass
    else:
        try:
            tmp = read_json_strings(os.path.join(_LOG_DIR, f))
            count += 1
            json_col.append(tmp)
        except json.decoder.JSONDecodeError:
            pass
count
```
